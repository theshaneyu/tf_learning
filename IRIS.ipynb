__這邊使用estimator作練習__

前處理

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./data/iris.csv')
data
```

```python
# 分data和label
X = data.iloc[:, :4]
y = data.iloc[:, 4]
print(X.shape)
print(y.shape)
```

```python
from keras.utils import np_utils
from sklearn.preprocessing import LabelEncoder
# 用於將label作one-hot-encoding
def label_data_onehot_encoding(label):
    # 將label encode成整數
    encoder = LabelEncoder()
    encoder.fit(label)
    encoded_label = encoder.transform(label) # 回傳一個numpy array，裡面是0, 1, 2 ...
    return encoded_label
```

```python
y = label_data_onehot_encoding(y)
y
```

```python
# 分training和testing資料
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# 全部都轉成numpy比較好操作
X_train = X_train.values
X_test = X_test.values
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

==============================================================================

開始建構computational graph (使用內建__estimator.DNNClassifier__)

宣告所有值都是numeric

```python
import tensorflow as tf
# 這邊不會用到column名字，所以key隨便亂設就好，主要是shape要是(4,)才會對應資料feature長度
# feature_columns = tf.feature_column.numeric_column('x', shape=(4,))
feature_columns = [tf.feature_column.numeric_column(key='x', shape=[4])]
```

宣告一個內建的__DNNClassifier__

```python
# model_dir參數是設定TensorFlow儲存check_point等等得地方(TensorBloard會用到)
classifier = tf.estimator.DNNClassifier(feature_columns=feature_columns,
                                        hidden_units=[10, 20, 10],
                                        n_classes=3,
                                        model_dir="./model_checkpoint/iris_model")
```

宣告__input_fn__（詳見筆記）

```python
train_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={'x':X_train},
    y=y_train,
    num_epochs=None,
    shuffle=True)
```

```python
classifier.train(input_fn=train_input_fn, steps=2000)
```

__evaluation__

```python
# Define the test inputs
test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": X_test},
    y=y_test,
    num_epochs=1,
    shuffle=False)
```

```python
# Evaluate accuracy.
accuracy_score = classifier.evaluate(input_fn=test_input_fn)["accuracy"]
```

predict 新進來的資料

```python
# 假設新進來的兩朵花如下
new_samples = np.array(
    [[6.4, 3.2, 4.5, 1.5],
     [5.8, 3.1, 5.0, 1.7]], dtype=np.float32)
# 寫一個predict的input_fn
predict_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"x": new_samples},
    num_epochs=1,
    shuffle=False)

predictions = list(classifier.predict(input_fn=predict_input_fn))
predicted_classes = [p["classes"] for p in predictions]

print(
    "New Samples, Class Predictions:    {}\n"
    .format(predicted_classes))
```
